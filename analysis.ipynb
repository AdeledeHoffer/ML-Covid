```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter


data=pd.read_csv('DATASET/England_complete.csv',index_col=False)
data=data.drop(columns='Unnamed: 0')
data['DATE']=pd.to_datetime(data['DATE'])
data=data[data['Month']>2]
mask=[True if 'X' not in seq else False for seq in data['Sequence']]
data=data[mask]

#con ripetizione
a1=data.groupby('Month').count()
a1

#con ripetizione
a2=data.groupby('Month').nunique()
table1=pd.DataFrame()
table1['Total Sequences']=a1['Sequence']
table1['Different sequences']=a2['Sequence']
table1.to_csv('ANALISI/Sequences_counts.csv')
print(table1)
table1=table1.reset_index()
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap

colors=['tab:blue','tab:orange','tab:green','tab:cyan','tab:olive','tab:purple','tab:red','tab:pink','tab:grey']

fig, ax = plt.subplots(figsize=(10,5))
for l in table1.columns[1:] :
    
    
    plt.plot(table1['Month'],table1[l],label=l,linewidth=2)
plt.title('Sequence distribution')
ax.set_xlabel('Months')
ax.set_ylabel('Counts ')
#ax.set_yscale('log')
plt.legend()
plt.savefig('ANALISI/sequences_distribution.png')
plt.show()

counters=[]
l_seq=[]
for k,v in data.groupby(['DATE']):
    cnt=0
    for el in v['Sequence'].values:
        if el not in l_seq:
            l_seq.append(el)
            cnt=cnt+1
    counters.append(cnt)
a=data.groupby('DATE').size().reset_index()
a['count']=counters
#number of new sequence
fig, ax= plt.subplots(figsize=(20,10))
plt.plot(a['DATE'],a['count'])
plt.show()

#analisi per lunghezze

fig, ax = plt.subplots(figsize=(20,10))
pal=sns.color_palette("viridis")[:3]
seq_len=list(map(lambda x: len(x),data['Sequence']))
data['leng']=seq_len



mask= list(map(lambda x: True if x in [1270,1273,1271] else False,data['leng'].values))
h=data[mask].groupby(['DATE','leng']).size().reset_index(name='counts')
h=h.drop([0,1])
sns.lineplot(data=h,x='DATE',y='counts',hue='leng',palette=sns.color_palette("husl", 3))
plt.savefig('ANALISI/length_dist.png')
plt.show()


h=data.groupby(by=['Month','lengths']).count()
mask= h['EPI ISL']>200
h=h[mask]
h=h.drop(columns=['EPI ISL','DATE','Sequence','Aligned'])
h=h.rename(columns={'ORIGIN LAB': 'counts'})
h.to_csv('ANALISI/lengths_analisi.csv')
h=h.reset_index()
#print(h)
#colors=['tab:blue','tab:orange','tab:green','tab:cyan','tab:olive','tab:purple','tab:red','tab:pink','tab:grey']
leng=list(set(h['lengths'].values))

fig, ax = plt.subplots(figsize=(15,8))

for i,l in enumerate(leng):
   
    mask=h['lengths']==l
    htemp=h[mask]
    plt.plot(htemp['Month'],htemp['counts'],label=l,linewidth=2)#,c=colors[i])
plt.title('Lengths distribution')
ax.set_xlabel('Months')
ax.set_ylabel('Counts (logarithmic scale)')
#ax.set_yscale('log')
plt.legend()
#plt.savefig('ANALISI/lengths_distribution.png')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter


data=pd.read_csv('DATASET/England_complete.csv')
diff_totale=[]
most_comm_seq=[]
for month in range(3,19):
    mask=data['Month']==month
    tmp=data[mask]
    h=tmp.groupby(['Sequence']).nunique().reset_index()
    diff_totale.append(len(h)/len(tmp))
    most_comm_seq.append(max(h['EPI ISL'])/len(tmp))
    
    
fig,ax=plt.subplots(figsize=(8,6))
plt.plot(range(3,19),diff_totale,label='Diff sequences/TOT sequences')
plt.plot(range(3,19),most_comm_seq,label='Most repeated sequence/t sequences')
plt.legend()
plt.xlabel('Months')
plt.ylabel('Percentage')
#plt.savefig('ANALISI/diff_sequences.png')
plt.show()
```
